```python
import numpy as np
import cv2
```

```python
image = cv2.imread('/content/shinchan.jpg')
```

```python
gray_image = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY) #convert color
edges = cv2.Canny(gray_image,50,100)
from google.colab.patches import cv2_imshow
cv2_imshow(image)
cv2_imshow(edges)
```
